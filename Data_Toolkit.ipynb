# 1. What is NumPy, and why is it widely used in Python?

NumPy (Numerical Python) is a powerful library for numerical computing in Python.
It provides support for large, multi-dimensional arrays and matrices along with a collection of mathematical functions to operate on these arrays.
It is widely used because it is efficient, fast, and forms the foundation for other libraries like Pandas, SciPy, and scikit-learn.

# 2. How does broadcasting work in NumPy?

Broadcasting allows NumPy to perform operations on arrays of different shapes.
It automatically expands the smaller array to match the shape of the larger array without copying data, enabling efficient and concise code for element-wise operations.


# 3. What is a Pandas DataFrame?

A Pandas DataFrame is a two-dimensional, labeled data structure with columns of potentially different types.
It is similar to a spreadsheet or SQL table and is widely used for data manipulation, cleaning, and analysis.


# 4. Explain the use of the groupby() method in Pandas.

The groupby() method is used to split the data into groups based on some criteria, apply a function to each group independently, and then combine the results.
It is commonly used for aggregation, transformation, and filtering operations.


# 5. Why is Seaborn preferred for statistical visualizations?

Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.
It is built on top of Matplotlib and integrates well with Pandas DataFrames, offering built-in themes, color palettes, and advanced plots like violin plots, heatmaps, and pair plots.


# 6. What are the differences between NumPy arrays and Python lists?

- NumPy arrays are more efficient for numerical operations and require less memory.
- They support vectorized operations, unlike Python lists.
- NumPy arrays are homogeneous, whereas lists can hold mixed data types.
- NumPy arrays provide more functionality for mathematical and logical operations.


# 7. What is a heatmap, and when should it be used?

A heatmap is a graphical representation of data where individual values are represented using color gradients.
It is useful for visualizing correlation matrices, missing data, and density distributions across two variables.


# 8. What does the term “vectorized operation” mean in NumPy?

Vectorized operations allow you to perform operations on entire arrays without writing explicit loops.
This results in more concise code and faster execution since the operations are implemented in optimized C code.

# 9. How does Matplotlib differ from Plotly?

- Matplotlib is a static plotting library, ideal for publication-quality visualizations.
- Plotly is an interactive plotting library that supports zooming, tooltips, and web-based sharing.
- Matplotlib is more customizable but less intuitive for interactivity, while Plotly offers built-in support for dashboards and interactive plots.


# 10. What is the significance of hierarchical indexing in Pandas?

Hierarchical indexing allows multiple levels of indexing in a DataFrame or Series.
It enables more complex data organization and facilitates operations like reshaping, slicing, and aggregation across multiple dimensions.


# 11. What is the role of Seaborn's pairplot() function?

The pairplot() function in Seaborn creates a grid of scatter plots for all pairwise combinations of numeric variables in a dataset, along with histograms or KDEs on the diagonal.
It is useful for exploring relationships and distributions in multivariate datasets.


# 12. What is the purpose of the describe() function in Pandas?

The describe() function provides a summary of descriptive statistics for numeric columns in a DataFrame, such as count, mean, standard deviation, min, max, and percentiles.
It helps in understanding the data distribution and spotting outliers.


# 13. Why is handling missing data important in Pandas?

Handling missing data is critical to ensure accurate analysis and modeling.
Unaddressed missing values can bias results, cause errors in computations, and degrade model performance.
Pandas provides methods like dropna(), fillna(), and interpolate() to handle such cases efficiently.


# 14. What are the benefits of using Plotly for data visualization?

- Offers interactive plots with zoom, hover, and tooltips.
- Built-in support for dashboards and web integration.
- Supports complex chart types like 3D plots, maps, and financial charts.
- Easy to use and integrates well with Pandas and Jupyter Notebooks.


# 15. How does NumPy handle multidimensional arrays?

NumPy handles multidimensional arrays using the ndarray object.
It supports arrays of arbitrary dimensions and provides operations like slicing, reshaping, transposing, and broadcasting to manipulate them efficiently.


# 16. What is the role of Bokeh in data visualization?

Bokeh is a powerful library for creating interactive and real-time web-based visualizations in Python.
It allows you to build complex dashboards and plots that can respond to UI events, making it ideal for big data and streaming data applications.


# 17. Explain the difference between apply() and map() in Pandas.

- apply() is used to apply a function along the axis of a DataFrame (rows or columns).
- map() is used for element-wise transformations and is typically applied to Series.
- apply() offers more flexibility, while map() is simpler and faster for 1D transformations.


# 18. What are some advanced features of NumPy?

- Broadcasting for efficient operations.
- Universal functions (ufuncs) for vectorized computations.
- Memory-mapped arrays for handling large datasets.
- Random number generation and statistical functions.
- Linear algebra, Fourier transforms, and polynomial functions.


# 19. How does Pandas simplify time series analysis?

Pandas offers built-in support for datetime objects, date ranges, resampling, shifting, rolling statistics, and time zone handling.
Its intuitive syntax and rich functionality make time series data manipulation and analysis seamless.


# 20. What is the role of a pivot table in Pandas?

Pivot tables summarize data by transforming or reshaping it, allowing for easy grouping and aggregation.
They are useful in generating multi-dimensional summaries and exploring relationships between different variables.


# 21. Why is NumPy's array slicing faster than Python's list slicing?

NumPy arrays are stored in contiguous memory blocks and use optimized C implementations for slicing.
Python lists, being objects with pointers, require more overhead.
As a result, NumPy slicing is faster and more efficient for large datasets.

- Exploring relationships using scatter plots, regression plots, and pair plots.
- Analyzing categorical data with bar plots and violin plots.

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

array_2d = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(array_2d, axis=1)
print("2D Array:\n", array_2d)
print("Sum of each row:", row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [78, 85, 92]})
mean_marks = df['Marks'].mean()
print("Mean of 'Marks' column:", mean_marks)
```

```python
# 3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 7, 2, 9, 4]
plt.scatter(x, y, color='blue')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns

df = pd.DataFrame({
    'Math': [90, 85, 88, 95, 70],
    'Science': [80, 78, 90, 85, 60],
    'English': [88, 82, 85, 90, 75]
})
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly.
import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Orange'], 'Quantity': [10, 15, 7]}
df_bar = pd.DataFrame(data)
fig = px.bar(df_bar, x='Fruits', y='Quantity', title='Fruit Quantity Bar Plot')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [50, 60, 70]})
df['Percentage'] = df['Marks'] / 100 * 100
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
product = np.multiply(a, b)
print("Element-wise multiplication:", product)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [1, 4, 2, 5, 3]
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
df = pd.DataFrame({'Student': ['A', 'B', 'C'], 'Score': [60, 85, 45]})
filtered_df = df[df['Score'] > 50]
print("Filtered DataFrame:\n", filtered_df)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
data = np.random.randn(1000)
sns.histplot(data, kde=True, bins=30, color='skyblue')
plt.title("Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy.
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7, 8]])
mat_product = np.dot(mat1, mat2)
print("Matrix multiplication result:\n", mat_product)
```

```python
# 13. Create a 3D scatter plot using Plotly.
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter3d(
    x=[1, 2, 3],
    y=[4, 5, 6],
    z=[7, 8, 9],
    mode='markers',
    marker=dict(size=8, color='blue')
)])
fig.update_layout(title='3D Scatter Plot', scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'
))
fig.show()
```
